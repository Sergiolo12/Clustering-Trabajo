```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# CARGAR DATASET
df = pd.read_csv('Mall_Customers.csv')
# Mostrar primeras filas
df.head()
```

```python
# Información general
df.info()

#Estadisticas descriptivas
df.describe()

# Valores nulos
df.isnull().sum()

# Distribución por género
df["Genre"].value_counts()
```

PROCESAMIENTO

```python
# Eliminación "Customer ID" por falta de aportación al análisis
df_clean = df.drop("CustomerID", axis=1)

# Encoding de la columna "Genre"
df_clean["Genre"] = df_clean["Genre"].map({"Male":0, "Female":1})

# Comprobación de la codificación
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_clean)
```

MÉTODO DEL CODO

```python
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Gráfico del método del codo
plt.figure(figsize=(8,4))
plt.plot(range(1,11), wcss, marker='o')
plt.title("Método del Codo")
plt.xlabel("Número Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

Análisis: El codo parece estar sobre 4 y 6 clusters, ya que luego se ve una bajada de WCSS

MÉTODO SILHOUETTE SCORE

```python
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_data)
    score = silhouette_score(scaled_data, labels)
    silhouette_scores.append(score)

# Gráfico de Silhouette Scores
plt.figure(figsize=(8,4))
plt.plot(range(2,11), silhouette_scores, marker="s", color="blue")
plt.title("Silhouette Score vs Número Clusters")
plt.xlabel("Número Clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

El mejor valor está en 10, pero se ve un crecimiento desde el 6 hasta el final

MODELO KMEANS

```python
kmeans_final = KMeans(n_clusters=5, random_state=42)
kmeans_final.fit(scaled_data)
df_clean["Cluster"] = kmeans_final.fit_predict(scaled_data)

# Gráfico Final
plt.figure(figsize=(9,5))
sns.scatterplot(data=df_clean, x="Annual Income (k$)", y="Spending Score (1-100)", hue="Cluster", palette="Set2")
plt.title("Clusters de clientes")
plt.xlabel("Ventas Globales Escaladas ")
plt.ylabel("Gastos (1-100)")
plt.legend()
plt.grid(True)
plt.show()
```

ANÁLISIS FINAL

```python
print("\nAnálisis de clusters:")
print(df_clean.groupby("Cluster").mean())
```

-Clúster 0 (Mayores, gasto promedio): 
Individuos predominantemente mayores (edad promedio ~56 años) con ingresos y gasto medios.
Este grupo tiene una proporción igualitaria de hombres y mujeres y se describe como clientes potencialmente conservadores, no grandes gastadores.

-Clúster 1 (Mujeres de alto ingreso, bajo gasto): 
Compuesto enteramente por mujeres con altos ingresos (promedio de $85k) pero con puntuaciones de gasto muy bajas.
Este segmento podría no estar interesado en los productos actuales o no percibir suficiente valor, presentando una oportunidad para campañas de fidelización o de lujo.

- Clúster 2 (Mujeres jóvenes, gasto alto):
Mujeres jóvenes (edad promedio 28 años) con ingresos medios ($61k) pero altos hábitos de gasto.
Son un segmento ideal para campañas de marketing dirigidas a jóvenes con capacidad y disposición de compra, adecuado para promociones y productos de moda.

-Clúster 3 (Hombres maduros, alto ingreso, gasto medio-alto):
Hombres maduros (edad promedio 38 años) con altos ingresos ($82k) y niveles de gasto aceptables.
Este es un segmento rentable y activo con poder adquisitivo, potencialmente receptivo al upselling o al asesoramiento experto.

-Clúster 4 (Hombres jóvenes, ingresos bajos-medios, buen gasto):
Hombres jóvenes (edad promedio 27 años) con ingresos bajos a medios ($39k) pero una puntuación de gasto decente.

Este grupo representa un segmento de consumidores masculinos jóvenes que, a pesar de ingresos moderados, muestran voluntad de gastar.
